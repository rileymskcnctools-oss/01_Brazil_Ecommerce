```python
import pandas as pd
sql_query = """
SELECT
    oord.review_score, 
    -- ÂÆûÈôÖÊÄªÁâ©ÊµÅËÄóÊó∂ÂùáÂÄº
    AVG(JULIANDAY(ood.order_delivered_customer_date) - JULIANDAY(ood.order_purchase_timestamp)) AS AVG_delivery_days,
    -- Âª∂ËøüÂ§©Êï∞ÂùáÂÄº
    AVG(JULIANDAY(ood.order_delivered_customer_date) - JULIANDAY(ood.order_estimated_delivery_date)) AS AVG_delay_days
FROM olist_orders_dataset ood
JOIN olist_order_reviews_dataset oord ON ood.order_id = oord.order_id 
WHERE ood.order_status = 'delivered'
  AND ood.order_delivered_customer_date IS NOT NULL 
  AND ood.order_estimated_delivery_date IS NOT NULL
GROUP BY 1 
ORDER BY oord.review_score DESC; -- Êîπ‰∏∫ÊåâËØÑÂàÜÊéíÂ∫èÔºåËßÇÂØüÂª∂ËøüÁöÑÂèòÂåñË∂ãÂäø
"""
df_logistic=pd.read_sql(sql_query,conn)
df_logistic.head()
```

```python
from config import *
# 2. SQL Êï∞ÊçÆÊèêÂèñ (Áâ©ÊµÅËÄóÊó∂‰∏éËØÑÂàÜÂÖ≥ËÅî)
query = """
SELECT
    oord.review_score, 
    AVG(JULIANDAY(ood.order_delivered_customer_date) - JULIANDAY(ood.order_purchase_timestamp)) AS AVG_delivery_days,
    AVG(JULIANDAY(ood.order_delivered_customer_date) - JULIANDAY(ood.order_estimated_delivery_date)) AS AVG_delay_days
FROM olist_orders_dataset ood
JOIN olist_order_reviews_dataset oord ON ood.order_id = oord.order_id 
WHERE ood.order_status = 'delivered'
  AND ood.order_delivered_customer_date IS NOT NULL 
GROUP BY 1 
ORDER BY 1 DESC;
"""

# Â∞ÜÁªìÊûúËØªÂÖ• DataFrame
df_logistic = pd.read_sql(query, conn)

# ==========================================
# 3. Ëá™Âä®ÂåñÂèØËßÜÂåñÁªòÂõæ
# ==========================================
plt.figure(figsize=(12, 7))
sns.set_style("whitegrid", {"font.sans-serif": [prop.get_name()]})

# --- ÁªòÂà∂‰∏ªËΩ¥ÔºöÂÆûÈôÖÊÄªÁâ©ÊµÅËÄóÊó∂ (Êü±Áä∂Âõæ) ---
ax1 = sns.barplot(
    x='review_score', 
    y='AVG_delivery_days', 
    data=df_logistic,
    palette='viridis_r',
    hue='review_score',  # ÊòéÁ°ÆÊåáÂÆöÊåâËØÑÂàÜÁùÄËâ≤
    legend=False           # ÈöêËóè‰æßËæπÂ§ö‰ΩôÁöÑÂõæ‰æã
)
ax1.set_title('‰∏ìÈ¢ò‰∫åÔºöÁâ©ÊµÅÊó∂ÊïàÂØπÁî®Êà∑ËØÑ‰ª∑ÁöÑÂΩ±ÂìçÂàÜÊûê', fontsize=16, pad=20)
ax1.set_xlabel('ËØÑ‰ª∑ÂæóÂàÜ (Review Score)', fontsize=12)
ax1.set_ylabel('Âπ≥ÂùáÊÄªÁâ©ÊµÅÂ§©Êï∞ (Â§©)', fontsize=12)

# --- ÁªòÂà∂ÂâØËΩ¥ÔºöÂπ≥ÂùáÂª∂ËøüÂ§©Êï∞ (ÊäòÁ∫øÂõæ) ---
ax2 = ax1.twinx()
sns.lineplot(
    x=range(len(df_logistic)), 
    y=df_logistic['AVG_delay_days'], 
    marker='o', 
    markersize=10,
    color='red', 
    ax=ax2, 
    linewidth=3, 
    label='Âπ≥ÂùáÂª∂Ëøü/ÊèêÂâçÂ§©Êï∞'
)
ax2.set_ylabel('Âπ≥ÂùáÂª∂ËøüÂ§©Êï∞ (Ë¥üÊï∞‰∏∫ÊèêÂâç)', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')

# --- Ê∑ªÂä†Êï∞ÂÄºÊ†áÊ≥® (ËÆ©ÂõæË°®Êõ¥ËÅå‰∏ö) ---
for i, v in enumerate(df_logistic['AVG_delivery_days']):
    ax1.text(i, v + 0.5, f'{v:.1f}Â§©', ha='center', fontweight='bold', color='darkblue')

# ÊòæÁ§∫Âõæ‰æãÂπ∂ÊâìÂç∞ÁªìËÆ∫
print("üìä ÂàÜÊûêÁªìËÆ∫:")
print(f"1. 5ÂàÜËÆ¢ÂçïÂπ≥ÂùáËÄóÊó∂ {df_logistic.iloc[0,1]:.1f} Â§©ÔºåËÄå1ÂàÜËÆ¢ÂçïËÄóÊó∂Ëææ {df_logistic.iloc[-1,1]:.1f} Â§©„ÄÇ")
print("2. Áâ©ÊµÅÊó∂ÈïøÁøªÂÄçÊòØÂØºËá¥Â∑ÆËØÑÁöÑÊ†∏ÂøÉËØ±Âõ†„ÄÇ")

plt.tight_layout()
plt.show()
```

```python
from config import *


# 1. ÊâßË°å SQL Ëé∑ÂèñÂú∞ÂüüÊï∞ÊçÆ
query_geo = """
SELECT 
    c.customer_state,
    COUNT(ood.order_id) AS total_orders,
    AVG(JULIANDAY(ood.order_delivered_customer_date) - JULIANDAY(ood.order_purchase_timestamp)) AS avg_delivery_days
FROM olist_orders_dataset ood
JOIN olist_customers_dataset c ON ood.customer_id = c.customer_id
WHERE ood.order_status = 'delivered'
  AND ood.order_delivered_customer_date IS NOT NULL
GROUP BY 1
HAVING total_orders > 50
ORDER BY avg_delivery_days DESC;
"""
df_geo = pd.read_sql(query_geo, conn)

# 2. ÁªòÂõæËÆæÁΩÆ
plt.figure(figsize=(12, 8))
# ‰ΩøÁî®Á∫¢Ëâ≤Ê∏êÂèòËâ≤ÔºåÈ¢úËâ≤Ë∂äÊ∑±‰ª£Ë°®Áâ©ÊµÅË∂äÊÖ¢ÔºåÈ¢ÑË≠¶ÊÑèÂë≥Ë∂äÂº∫
ax = sns.barplot(
    x='avg_delivery_days', 
    y='customer_state', 
    data=df_geo.head(15), # Â±ïÁ§∫ÊúÄÊÖ¢ÁöÑÂâç15‰∏™Â∑û
   hue='customer_state',  # ÊòéÁ°ÆÊåáÂÆöÊåâÂ∑ûÂêçÁùÄËâ≤
    palette='Reds_r',
    legend=False           # ÈöêËóè‰æßËæπÂ§ö‰ΩôÁöÑÂõæ‰æã
)

# 3. Ê∑ªÂä†Ë£ÖÈ•∞‰∏éÊ†áÊ≥®
plt.title('‰∏ìÈ¢ò‰∏âÔºöÂ∑¥Ë•øÂêÑÂ∑ûÁâ©ÊµÅËÄóÊó∂ÊéíË°åÊ¶ú (Top 15 ÊúÄÊÖ¢Âú∞Âå∫)', fontsize=16, pad=20)
plt.xlabel('Âπ≥ÂùáÈÄÅËææÂ§©Êï∞ (Â§©)', fontsize=12)
plt.ylabel('Â∑ûÁº©ÂÜô (State Code)', fontsize=12)

# Âú®Êù°ÂΩ¢ÂõæÊú´Á´ØÂä†‰∏äÂÖ∑‰ΩìÂ§©Êï∞
for p in ax.patches:
    ax.annotate(f'{p.get_width():.1f}Â§©', 
                (p.get_width() + 0.2, p.get_y() + p.get_height()/2), 
                va='center', fontsize=10)

# Ê∑ªÂä†‰∏ÄÊù°ÂÖ®Âπ≥Âè∞Âπ≥ÂùáÁ∫øÁöÑÂèÇËÄÉÁÇπÔºàÂÅáËÆæÂÖ®Âπ≥Âè∞Âπ≥Âùá12Â§©Ôºâ
plt.axvline(12.5, color='blue', linestyle='--', label='ÂÖ®Âπ≥Âè∞Âπ≥ÂùáÊó∂Êïà (12.5Â§©)')
plt.legend()

plt.tight_layout()
plt.show()
```
