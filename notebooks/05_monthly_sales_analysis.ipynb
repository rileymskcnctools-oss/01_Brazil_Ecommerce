```python
# 模仿要点：使用双轴图 (Dual Axis) 对比不同量级的指标
# 读取数据
with open('../sql_scripts/05_monthly_sales_fluctuation.sql', 'r') as f:
    df_sales = pd.read_sql(f.read(), conn)

# 绘图逻辑
fig, ax1 = plt.subplots(figsize=(14, 7))


# 1. 绘制柱状图 (左轴：订单量)
sns.barplot(x='month_idx', y='order_count', data=df_sales, ax=ax1, alpha=0.5, color='royalblue')
ax1.set_ylabel('月订单量', fontsize=12)
ax1.tick_params(axis='x', rotation=45)

# 2. 绘制折线图 (右轴：销售额)
ax2 = ax1.twinx()
sns.lineplot(x='month_idx', y='total_revenue', data=df_sales, ax=ax2, marker='o', color='red', linewidth=2)
ax2.set_ylabel('月销售额 (GMV)', fontsize=12)

plt.title('专题五：月度销售趋势分析（流量与营收对比）', fontsize=16)
plt.show()
```
